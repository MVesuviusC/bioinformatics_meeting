```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import colorcet as cc
```

```python
mpg = sns.load_dataset(
    "mpg"
)
mpg
```

```python
plt.scatter(
    "displacement",
    "mpg",
    data= mpg
)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(121)
ax.scatter(
    "displacement",
    "mpg",
    data = mpg
)
ax.set_xlabel("Displacement (cc)")
ax.set_ylabel("MPG")
ax = fig.add_subplot(122, sharey= ax)
ax.scatter(
    "horsepower",
    "mpg",
    data= mpg
)
ax.set_xlabel("Horsepower")
```

```python
fig, axes = plt.subplots(1, 2, sharey= "row")
axes[0].set_ylabel("MPG")
for ax, c_name in zip(axes, ["displacement", "horsepower"]):
    ax.scatter(
        c_name,
        "mpg",
        data= mpg
    )
    if c_name == "displacement":
        c_name += " (cc)"
    ax.set_xlabel(c_name.capitalize())
fig.tight_layout()
```

```python
mapping = {ori: c for ori, c in zip(mpg["origin"].unique(), cc.glasbey_category10)}
mapping
```

```python
fig, ax = plt.subplots()
for origin, color in mapping.items(): 
    ax.scatter(
        "displacement",
        "mpg",
        color= color,
        label= origin.capitalize(),
        data= mpg.loc[mpg["origin"] == origin]
    )
ax.legend(title= "origin")
ax.set_xlabel("Displacement (cc)")
ax.set_ylabel("MPG")
```

```python
ax = sns.scatterplot(
    data= mpg,
    x= "displacement",
    y= "mpg",
    hue= "origin"
)
ax.set_xlabel("Displacement (cc)")
ax.set_ylabel("MPG")
ax.legend(title= "Origin")
```

```python
fig, axes = plt.subplots(
    1, 
    2, 
    sharey= "row",
    figsize= (10, 5),
)
for ax, metric in zip(axes, ["displacement", "horsepower"]):
    sns.scatterplot(
        data= mpg,
        x= metric,
        y= "mpg",
        hue= mpg["origin"].str.capitalize(),
        ax= ax,
        alpha= 0.8
    )
    ax.legend(title= "Origin")
    ax.set_xlabel(metric.capitalize())
axes[0].set_ylabel("MPG")
fig.tight_layout()
```
